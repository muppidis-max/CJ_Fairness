## Load dataset

```python
#@title Create a dataframe that contains the metadata and condition labels

from google.cloud import storage

def list_blobs(storage_client, bucket_name):
  """Helper to list blobs in a bucket (useful for debugging)."""
  blobs = storage_client.list_blobs(bucket_name)
  for blob in blobs:
    print(blob)

def initialize_df_with_metadata(bucket, csv_path):
  """Loads the given CSV into a pd.DataFrame."""
  df = pd.read_csv(io.BytesIO(bucket.blob(csv_path).download_as_string()), dtype={'case_id': str})
  df['case_id'] = df['case_id'].astype(str)
  return df

def augment_metadata_with_labels(df, bucket, csv_path):
  """Loads the given CSV into a pd.DataFrame."""
  labels_df = pd.read_csv(io.BytesIO(bucket.blob(csv_path).download_as_string()), dtype={'case_id': str})
  labels_df['case_id'] = labels_df['case_id'].astype(str)
  merged_df = pd.merge(df, labels_df, on='case_id')
  return merged_df

Globals.gcs_storage_client = storage.Client(Globals.gcp_project)
Globals.gcs_bucket = Globals.gcs_storage_client.bucket(
    Globals.gcs_bucket_name
)
Globals.cases_df = initialize_df_with_metadata(Globals.gcs_bucket, Globals.cases_csv)
Globals.cases_and_labels_df = augment_metadata_with_labels(Globals.cases_df, Globals.gcs_bucket, Globals.labels_csv)
print(len(Globals.cases_and_labels_df))
```

## Get familiar with the data

```python
#@title Display the images for a case (and condition labels, optionally)
import random

import io
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def display_image(bucket, image_path):
  image = Image.open(io.BytesIO(bucket.blob(image_path).download_as_string()))
  figure_size=4
  f, axarr = plt.subplots(1, 1, figsize = (figure_size, figure_size))
  axarr.imshow(image, cmap='gray')
  axarr.axis('off')
  plt.show()

def display_images_for_case(df, case_id="", print_condition_labels=True):
  # Use a random case if none is provided:
  if case_id:
    matched_df = df[df['case_id'] == case_id]
  else:
    matched_df = df.sample(1)

  image_paths = matched_df[Globals.image_path_columns].values.tolist()[0]
  for path in image_paths:
    if isinstance(path, str):
      display_image(Globals.gcs_bucket, path)
  if print_condition_labels:
    condition_labels = matched_df[[Globals.weighted_skin_condition_label]].values.tolist()[0]
    for label in condition_labels:
      if isinstance(label, str):
        print(label)

# display_images_for_case(Globals.cases_and_labels_df, "-1000600354148496558")
display_images_for_case(Globals.cases_and_labels_df)
```

```python
#@title Print metadata stats
def print_metadata_stats(df: pd.DataFrame) -> None:
  print(f'Num cases: {len(df)}')

  # Number of cases with 1, 2, or 3 images
  image_1_cnt = df["image_1_path"].count()
  image_2_cnt = df["image_2_path"].count()
  image_3_cnt = df["image_3_path"].count()
  print(f'Num cases with image_1_path set: {image_1_cnt}')
  print(f'Num cases with image_2_path set: {image_2_cnt}')
  print(f'Num cases with image_3_path set: {image_3_cnt}')
  print(f'Total images: {image_1_cnt + image_2_cnt + image_3_cnt}')

  # Sex at birth:
  print('\nSelf-reported Sex at Birth:')
  sab = collections.Counter(df['sex_at_birth'].dropna())
  total = sum(sab.values())
  for k in sorted(sab):
    print(f'  {k} {sab[k]} ({100.0*sab[k]/total:.1f}%)')

  # FST distribution:
  print('\nSelf-reported Fitzpatrick Skin Type:')
  fst = collections.Counter(df['fitzpatrick_skin_type'].dropna())
  total = sum(fst.values())
  for k in sorted(fst):
    print(f'  {k} {fst[k]} ({100.0*fst[k]/total:.1f}%)')


print_metadata_stats(Globals.cases_and_labels_df)
```

```python
#@title Add combined_race_ethnicity column

def merge_race_ethnicity_columns(row):
  combined = []
  for col in row.keys():
    if col.startswith("race_ethnicity_") and row[col] == 'YES':
      combined.append(col[len("race_ethnicity_"):])
  return ",".join(combined)

Globals.cases_and_labels_df["combined_race_ethnicity"] = Globals.cases_and_labels_df.apply(merge_race_ethnicity_columns, axis=1)
Globals.cases_and_labels_df[["case_id", "combined_race_ethnicity"]]
```

```python
#@title Print race/ethnicity distribution

def print_race_ethnicity_dist(df):
  c = collections.Counter(df['combined_race_ethnicity'].dropna())
  total = sum(c.values())
  for k in sorted(c):
    if not k:
      k = 'unspecified'
    print(f'  {k}: {c[k]} ({100.0*c[k]/total:.1f}%)')

print_race_ethnicity_dist(Globals.cases_and_labels_df)
```

## Display distributions of self-reported condition info

```python
#@title Display body part distribution

def get_distribution_columns(df, prefix):
  cols = [c for c in df.columns if c.startswith(prefix)]
  dist = collections.Counter()
  for c in cols:
    name = c[len(prefix):]
    val = sum(df[c] == "YES")
    dist[name] = val
  return dist

def get_distribution_and_display(df, prefix):
  ctr = get_distribution_columns(df, prefix)
  dist = dict(ctr.most_common())
  print(f'Distribution data: {dist}')
  names = list(dist.keys())
  counts = list(dist.values())
  plt.bar(names, counts)
  plt.xticks(rotation=90)  # Rotate labels by 90 degrees
  plt.show()

# Distribution data: {'arm': 1707, 'leg': 1694, 'torso_front': 795, 'head_or_neck': 737, 'torso_back': 652, 'other': 543, 'back_of_hand': 521, 'foot_top_or_side': 415, 'buttocks': 391, 'genitalia_or_groin': 222, 'palm': 215, 'foot_sole': 104}
get_distribution_and_display(Globals.cases_and_labels_df, "body_parts_")
```

# Task
Implement a basic online augmentation strategy that includes setting up a `tf.data.Dataset` from the image paths and labels in `Globals.cases_and_labels_df`, creating a function for basic geometric and color augmentations (e.g., `tf.image.random_flip_left_right`, `tf.image.random_brightness`, `tf.image.random_contrast`, `tf.image.random_hue`, `tf.image.random_saturation`), and visualizing a batch of augmented images from the pipeline.

```python
import tensorflow as tf
import collections
import matplotlib.pyplot as plt
import io
import numpy as np

# 1. Prepare Labels
# 1a. Extract all unique skin condition names
all_conditions = set()
for entry in Globals.cases_and_labels_df[Globals.weighted_skin_condition_label].dropna():
  try:
    conditions_dict = eval(entry)
    all_conditions.update(conditions_dict.keys())
  except (NameError, SyntaxError) as e:
    print(f"Error parsing label string '{entry}': {e}")

unique_conditions = sorted(list(all_conditions))

# 1b. Create a dictionary mapping each unique condition name to a unique integer index.
condition_to_idx = {condition: i for i, condition in enumerate(unique_conditions)}
idx_to_condition = {i: condition for condition, i in condition_to_idx.items()}
num_conditions = len(unique_conditions)
print(f"Total unique conditions: {num_conditions}")

def _parse_labels_py(label_string_tensor):
  """Converts a label string (dict representation) to a multi-hot encoded tensor in Python."""
  # label_string_tensor is expected to be a tf.Tensor of type string.
  # We need to convert it to a Python bytes object before decoding and eval'ing.
  label_string_np_bytes = label_string_tensor.numpy()
  label_dict = eval(label_string_np_bytes.decode('utf-8'))

  multi_hot_encoded = np.zeros(num_conditions, dtype=np.float32)
  for condition, weight in label_dict.items():
    if condition in condition_to_idx:
      multi_hot_encoded[condition_to_idx[condition]] = float(weight)
  return multi_hot_encoded

def parse_labels(label_string):
  """Wrapper for _parse_labels_py to be used with tf.data.Dataset.map."""
  multi_hot_encoded_tensor = tf.py_function(
      _parse_labels_py, [label_string], tf.float32)
  multi_hot_encoded_tensor.set_shape([num_conditions])
  return multi_hot_encoded_tensor

# 2. Filter Data and Create Dataset
# 2a. Filter Globals.cases_and_labels_df
filtered_df = Globals.cases_and_labels_df[
    Globals.cases_and_labels_df['image_1_path'].notnull() &
    Globals.cases_and_labels_df[Globals.weighted_skin_condition_label].notnull()
]

# 2b. Create a tf.data.Dataset
image_paths = filtered_df['image_1_path'].values
label_strings = filtered_df[Globals.weighted_skin_condition_label].values
dataset = tf.data.Dataset.from_tensor_slices((image_paths, label_strings))

# Get IMG_HEIGHT and IMG_WIDTH from kernel state (or define if not present)
IMG_HEIGHT = 256 # Assume 256 based on kernel state
IMG_WIDTH = 256  # Assume 256 based on kernel state

# 3. Define Image Processing and Augmentation Functions
def _load_image_py(image_path_tensor):
  """Loads image in Python, handles errors, returns NumPy array."""
  image_path_str = image_path_tensor.numpy().decode('utf-8')
  gcs_path = f"gs://{Globals.gcs_bucket_name}/{image_path_str}"
  try:
    with tf.io.gfile.GFile(gcs_path, 'rb') as f:
        img_bytes = f.read()

    img = tf.image.decode_image(img_bytes, channels=3) # Use decode_image for general format
    img = tf.image.resize(img, [IMG_HEIGHT, IMG_WIDTH])
    img = tf.cast(img, tf.float32) / 255.0
    return img.numpy() # Return as numpy array for tf.py_function
  except (tf.errors.NotFoundError, tf.errors.InvalidArgumentError, Exception) as e: # Corrected typo here
    print(f"Warning: Error loading image {gcs_path}. {type(e).__name__}: {e}. Returning a blank image.")
    return np.zeros((IMG_HEIGHT, IMG_WIDTH, 3), dtype=np.float32)

def load_image(image_path):
  """Wrapper for _load_image_py to be used with tf.data.Dataset.map."""
  image_tensor = tf.py_function(
      _load_image_py, [image_path], tf.float32)
  image_tensor.set_shape([IMG_HEIGHT, IMG_WIDTH, 3])
  return image_tensor

def augment_image(image, label):
  """Applies random geometric and color augmentations to the image."""
  image = tf.image.random_flip_left_right(image)
  image = tf.image.random_brightness(image, max_delta=0.2)
  image = tf.image.random_contrast(image, lower=0.8, upper=1.2)
  image = tf.image.random_hue(image, max_delta=0.05)
  image = tf.image.random_saturation(image, lower=0.8, upper=1.2)
  return image, label

# 4. Build and Process the Data Pipeline
batch_size = 8 # Example batch size

# Map load_image and parse_labels
dataset = dataset.map(
    lambda img_path, lbl_str: (load_image(img_path), parse_labels(lbl_str)),
    num_parallel_calls=tf.data.AUTOTUNE
)

# Apply augmentation
augmented_dataset = dataset.map(augment_image, num_parallel_calls=tf.data.AUTOTUNE)

# Shuffle, batch, prefetch
augmented_dataset = augmented_dataset.shuffle(buffer_size=1000).batch(batch_size).prefetch(tf.data.AUTOTUNE)

print("Dataset pipeline created successfully.")

# 5. Visualize Augmented Images
fig, axes = plt.subplots(batch_size // 4 + (batch_size % 4 > 0), 4, figsize=(20, 5 * (batch_size // 4 + (batch_size % 4 > 0))))
axes = axes.flatten()

for i, (images, labels) in enumerate(augmented_dataset.take(1)):
  for j in range(images.shape[0]):
    ax = axes[j]
    ax.imshow(images[j].numpy()) # Images are already normalized to [0,1]
    ax.axis('off')

    # Get top 3 condition labels and weights for display
    current_label_tensor = labels[j].numpy()
    present_conditions_with_weights = []
    for cond_idx, weight in enumerate(current_label_tensor):
      if weight > 0:
        present_conditions_with_weights.append((idx_to_condition[cond_idx], weight))

    # Sort by weight in descending order
    present_conditions_with_weights.sort(key=lambda x: x[1], reverse=True)

    title_text = "\n".join([f"{cond}: {weight:.2f}" for cond, weight in present_conditions_with_weights[:3]])
    ax.set_title(title_text, fontsize=10)

plt.tight_layout()
plt.show()

print("Visualization of augmented images complete.")
```
